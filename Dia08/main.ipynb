```python
puntuacion = []

for i in range (0, len (matriz)):
    for j in range (0, len (matriz)):
        fila = i
        columna = j
        # Vamos a ver cuantos arboles podemos llegar a ver
        right = 0
        left = 0
        up = 0
        down = 0
        if fila == 0 or columna == 0 or fila == len(matriz) - 1 or columna == len(matriz) - 1:
            puntuacion.append(0)
        else:
            print("Fila: ", fila, "Columna: ", columna)
            for k in range (columna + 1, len(matriz)):
                if matriz[fila][columna] > matriz[fila][k]:
                    right += 1
                if matriz[fila][columna] == matriz[fila][k] or matriz[fila][columna] < matriz[fila][k]:
                    right += 1
                    break
            for k in range (columna - 1, -1, -1):
                if matriz[fila][columna] > matriz[fila][k]:
                    left += 1
                if matriz[fila][columna] == matriz[fila][k] or matriz[fila][columna] < matriz[fila][k]:
                    left += 1
                    break
            for k in range (fila + 1, len(matriz)):
                if matriz[fila][columna] > matriz[k][columna]:
                    down += 1
                if matriz[fila][columna] == matriz[k][columna] or matriz[fila][columna] < matriz[k][columna]:
                    down += 1
                    break
            for k in range (fila - 1, -1, -1):
                if matriz[fila][columna] > matriz[k][columna]:
                    up += 1
                if matriz[fila][columna] == matriz[k][columna] or matriz[fila][columna] < matriz[k][columna]:
                    up += 1
                    break
            print("Right: ", right, "Left: ", left, "Up: ", up, "Down: ", down)
            print("Puntuacion: ", right * left * up * down)
            puntuacion.append( right * left * up * down)
            
print(max(puntuacion))
```
